```python
plt.figure(figsize = (6,6))

# KDE for males
sns.kdeplot(data=df[df["gender"] == "male"], x="reading score", y="math score", cmap="Blues", fill=True, alpha=0.5, label="Male")

# KDE for females
sns.kdeplot(data=df[df["gender"] == "female"], x="reading score", y="math score", cmap="Reds", fill=True, alpha=0.5, label="Female")

plt.title("KDE Plot: math score vS Sex")
plt.legend()
plt.show()
```

```python
figbi , axesbi = plt.subplots(2,3,figsize = (16,10))

student.groupby("race/ethnicity")["math score"].mean().plot(kind = "barh" ,ax = axesbi[0,0] ,  xlim = [0,100])
student.groupby("race/ethnicity")["reading score"].mean().plot(kind = "barh" , ax = axesbi[0,1], xlim = [0,100])
student.groupby("race/ethnicity")["writing score"].mean().plot(kind = "barh" , ax = axesbi[0,2], xlim = [0,100])

```
