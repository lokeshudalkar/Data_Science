```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#To render plot in noteBook
%matplotlib inline
```

```python
x = [1,4,8,16 ,25]
y = [1,5,10,15,20]
```

```python
fig = plt.figure(figsize=(10,3))
plt.plot(x,y)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Line Plot")
plt.suptitle("Demo Plot",size = 20 , y = 1.03)
plt.show()
```

```python
#Modification in above command
plt.hist(mtCars["mpg"] , facecolor = "k" , edgecolor = "b", bins = 5)
```

```python
#BoxPlot
plt.boxplot(mtCars["mpg"] , vert = True)
```

```python
#BoxPlot for vertical view
plt.boxplot(mtCars["mpg"] , vert = False)
plt.xlabel("Mpg-Values")
plt.ylabel("Mpg")
plt.title("Boxplot")
```
