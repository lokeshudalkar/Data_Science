#tips visulaization

```python
#seaborn code from here
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
tips.shape
```

```python
#Strip plot
# sns.stripplot(y = "tip" , data = tips , jitter = False)
# plt.ylabel("tip ($) ")
# plt.show()
# above code will show values in one line becuase jitter is false
sns.stripplot(y = "tip" , data = tips , jitter = True)
plt.ylabel("tip ($) ")
plt.show()
```

```python
#Strip plot with grouping columns
sns.stripplot(x = "day" , y = "tip" , data = tips , jitter = False)
plt.ylabel("tip ($) ")
plt.show()
```

```python
sns.stripplot(x = "day" , y = "tip" , data = tips , jitter = True)
plt.xlabel("Days of week")
plt.ylabel("tip ($) ")
plt.show()
```

```python
#swam plot
sns.swarmplot(x = 'day' , y = 'tip' , data = tips)
plt.xlabel("Days of week")
plt.ylabel("tip ($) ")
plt.show()
```

```python
#swam plot showing according to theire gender
sns.swarmplot(x = 'day' , y = 'tip' , data = tips , hue="sex")
plt.xlabel("Days of week")
plt.ylabel("tip ($) ")
plt.show()
#hue refers to color
```

```python
#changing orientation
sns.swarmplot(x = 'tip' , y = 'day' , data = tips , hue='sex' , orient='h')
plt.ylabel("Days of week")
plt.xlabel("tip ($) ")
plt.show()
```

##Box and violine plot


```python
sns.boxplot(x = "day" , y = "tip" , data = tips , palette = "Set2")
plt.ylabel("tip ($) ")
```

```python
sns.violinplot(x = "day" , y = "tip" , data = tips , palette = "Set2")
plt.ylabel("tip ($) ")
plt.show()
```

```python
#printing both in same cell
plt.subplot(1,2,1)
sns.boxplot(x = "day" , y = "tip" , data = tips , palette = "Set2")
plt.ylabel("tip ($) ")
plt.subplot(1,2,2)
sns.violinplot(x = "day" , y = "tip" , data = tips , palette = "Set2")
plt.ylabel("tip ($) ")
plt.show()
```

```python
#both combined in same plot
sns.violinplot(x = "day" , y = "tip" , data = tips , color = "lightgray" , inner = None)
sns.stripplot(x = "day" , y = "tip" , data = tips , jitter = True , size = 4 ,  palette = "Set2")
plt.ylabel("tip ($) ")
plt.show()
```

##Bivariate visualization -> two variable data visualization


```python
#joint plot
sns.jointplot(x = "total_bill" , y = "tip" , data = tips)
```

```python
#density plot
sns.jointplot(x = "total_bill" , y = "tip" , data = tips , kind = "kde")
#kde is kernal density
```

```python
#pair plot
sns.pairplot(tips)
```

```python
sns.pairplot(tips , hue = "sex")
```

```python
survived = df.loc[df["Survived"]== 1 , "Age"].dropna()
sns.distplot(survived)
plt.title("Survived")
```

```python
not_survived = df.loc[df["Survived"]== 0 , "Age"].dropna()
sns.distplot(not_survived)
plt.title("Not Survived")
```

```python
sns.pairplot(df.dropna())
```

```python
sns.pairplot(df.dropna() , hue= "Sex")
```

```python
sns.pairplot(df.dropna() , hue = "Survived")
```

```python
grid = sns.FacetGrid(df  , col = "Survived", row = "Pclass" , height = 2.4 , aspect = 1.5 )
grid.map(plt.hist , "Age" , alpha = .5 , bins = 20)
grid.add_legend()
```

```python
sns.distplot(a=df[df['Embarked']== 'C']['Survived'], bins=3, kde=False)
plt.title("Cherbourg")
plt.xticks([0, 1])

plt.show()
sns.distplot(a=df[df['Embarked']== 'Q']['Survived'], bins=3, kde=False)
plt.title("QueensTown")
plt.xticks([0, 1])

plt.show()
sns.distplot(a=df[df['Embarked']== 'S']['Survived'], bins=3, kde=False)
plt.title("Southampton")
plt.xticks([0, 1])
```

```python
sns.displot(data=df[df['Embarked'] == 'S'], x='Survived', bins=2, kde=False , hue ='Sex')
```

```python
sns.jointplot(x = "Age" , y = "Survived" , data = df , kind = "kde")
```

```python
sns.jointplot(data=df, x="Age", y="Survived", kind="reg")
```

```python
plt.figure(figsize = (6,6))

# KDE for males
sns.kdeplot(data=df[df["Sex"] == "male"], x="Age", y="Survived", cmap="Blues", fill=True, alpha=0.5, label="Male")

# KDE for females
sns.kdeplot(data=df[df["Sex"] == "female"], x="Age", y="Survived", cmap="Reds", fill=True, alpha=0.5, label="Female")

plt.title("KDE Plot: Age vs Survived by Sex")
plt.legend()
plt.show()
```
