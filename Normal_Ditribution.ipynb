#Importing statements

```python
import numpy as np
import pandas as pd
from scipy import stats
```

#Importing CSv Files


```python
beml_df = pd.read_csv('/content/BEML.csv')
```

```python
beml_df
```

```python
glaxo_df = pd.read_csv('/content/GLAXO.csv')
```

```python
glaxo_df
```

#Computing Gain


```python
#for we only need date and closing price
beml = beml_df[['Date','Close']]
glaxo = glaxo_df[['Date','Close']]
```

```python
beml = beml.set_index(pd.DatetimeIndex(beml['Date']))
glaxo = glaxo.set_index(pd.DatetimeIndex(glaxo['Date']))
```

```python
beml.head()
```

```python
glaxo.head()
```

#visualisation of beml & glaxo

```python
#For computing Gain we need to visualize data first
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
plt.plot(beml["Close"])
plt.xlabel("date")
plt.ylabel("close")
```

```python
plt.plot(glaxo["Close"])
plt.xlabel("date")
plt.ylabel("close")
```

```python
beml["gain"] = beml["Close"].pct_change(periods = 1)
glaxo["gain"] = glaxo["Close"].pct_change(periods = 1)
beml  = beml.dropna()
glaxo = glaxo.dropna()
```

```python
beml
```

```python
glaxo
```

```python
#plot gains
plt.figure(figsize=(8,6))
plt.plot(beml["gain"])
plt.xlabel("date")
plt.ylabel("Gain")
```

```python
#plot gains
plt.figure(figsize=(8,6))
plt.plot(glaxo["gain"])
plt.xlabel("date")
```

```python
sns.distplot(glaxo.gain , label = "Glaxo")
plt.xlabel("gain")
plt.ylabel("density")
plt.legend()
```

```python
sns.displot(beml.gain , label = "Beml")
plt.xlabel("gain")
plt.ylabel("density")
plt.legend()
```

```python
print("Mean : " , round(beml.gain.mean(),4))
print("Standard Deviation : " , round(beml.gain.std(),4))
```

```python
#probability of making 2% or higher loss
stats.norm.cdf(-0.02 , loc = beml.gain.mean() , scale = beml.gain.std())
```

```python
#probability of making 2% or higher gain
1 - stats.norm.cdf(0.02 , loc = beml.gain.mean() , scale = beml.gain.std())
```

```python
stats.norm.cdf(-0.02 , loc = glaxo.gain.mean() , scale = glaxo.gain.std())
```

```python
1 - stats.norm.cdf(0.02 , loc = glaxo.gain.mean() , scale = glaxo.gain.std())
```

#Confidence Interval

```python
glaxo_df_ci = stats.norm.interval(0.95,
loc = glaxo.gain.mean(),
scale = glaxo.gain.std())
print("Gain at 95% confidence interval is : " , np.round(glaxo_df_ci,4))
```

```python
beml_df_ci = stats.norm.interval(0.95,
loc = beml.gain.mean(),
scale = beml.gain.std())
print("Gain at 95% confidence interval is : " , np.round(beml_df_ci,4))
```
