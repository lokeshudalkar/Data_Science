#Computing Gain


```python
#for we only need date and closing price
beml = beml_df[['Date','Close']]
glaxo = glaxo_df[['Date','Close']]
```

```python
beml = beml.set_index(pd.DatetimeIndex(beml['Date']))
glaxo = glaxo.set_index(pd.DatetimeIndex(glaxo['Date']))
```

#visualisation of beml & glaxo

```python
#For computing Gain we need to visualize data first
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
plt.plot(beml["Close"])
plt.xlabel("date")
plt.ylabel("close")
```

```python
plt.plot(glaxo["Close"])
plt.xlabel("date")
plt.ylabel("close")
```

```python
beml["gain"] = beml["Close"].pct_change(periods = 1)
glaxo["gain"] = glaxo["Close"].pct_change(periods = 1)
beml  = beml.dropna()
glaxo = glaxo.dropna()
```
